# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
import numpy as np
import matplotlib as plt
url = "https://raw.githubusercontent.com/USMS-FPK-IA-GI-2025/controle-01-numpy-and-matplotlib-iris-Bouchask/main/data/Iris.csv"
data = np.genfromtxt(url ,delimiter="," , skip_header=1 ,dtype = str )
print(data.shape)
print(data)
```

# Partie B — Statistiques Basiques et Opérations Mathématiques

### B6. Moyenne par colonne
Calculez la moyenne de chaque colonne de `features` avec `np.mean(axis=0)`.


```python
X = X.astype(float)
mean_features = np.mean(X , axis = 0)
print(mean_features)
```

### B7. Min et Max
Trouvez la valeur minimale et maximale de la longueur des pétales.


```python
max_feature_petal = np.max(X[:,3] , axis = 0)
min_feature_petal = np.min(X[:,3] ,axis = 0)
print(max_feature_petal)
print(min_feature_petal)
```

### B9. Normalisation
Normalisez les longueurs des pétales en les divisant par leur maximum.
Stockez le résultat dans `normalized_petals`.


```python
normalized_petals = X[:,3] / max_feature_petal
print(normalized_petals[:5])
```

# Partie D — Visualisation avec Matplotlib (Plots Simples)

### D17. Line plot
Line plot des longueurs des sépales pour les 50 premières fleurs.


```python
# Votre code ici
import matplotlib.pyplot as plt

y_value = X_clean_id[:50,0]
plt.plot(y_value)
plt.title("Line plot des longueurs des sépales pour les 50 premières fleurs")
plt.xlabel("index de la fleur")
plt.ylabel("sepal length (cm)")
plt.show()
```

### D18. Scatter plot
Scatter plot : longueur vs largeur des pétales.


```python
# Votre code ici
x_value = X_clean_id[:,2]
y_value = X_clean_id[:,3]
plt.scatter(x_value,y_value, c="blue" , marker="o",alpha=0.6)
plt.title("Scatter plot : longueur vs largeur des pétales")
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.show()
```

### D19. Bar chart
Bar chart des moyennes des quatre caractéristiques.


```python
# Votre code ici
mean_value = np.mean(X_clean_id , axis=0)
features = ["SepalLengthCm",	"SepalWidthCm",	"PetalLengthCm",	"PetalWidthCm"]
plt.bar(features , mean_value , color = ['green', 'blue', 'orange', 'red'] )
plt.title("Bar chart des moyennes des quatre caractéristiques")
plt.ylabel("mean features")
plt.xlabel("features")
plt.show()
```

### D20. Histogram
Histogramme de la largeur des sépales (10 bins).


```python
# Votre code ici
sepal_width = X_clean_id[:,1]
plt.hist(sepal_width , bins=10 , color = "blue" , alpha = 0.7 ,edgecolor = "black")
plt.title("Histogramme de la largeur des sépales (10 bins)")
plt.xlabel("sepal width")
plt.ylabel("frequence")
plt.show()
```

### D21. Boxplot par espèce
Boxplot des longueurs des pétales par espèce.


```python
# Votre code ici
setosa  = X_clean_id[Y == "Iris-setosa" , 2]
versicolor = X_clean_id[Y == "Iris-versicolor" , 2]
virginica = X_clean_id[Y == "Iris-virginica" , 2 ]
plt.figure(figsize=(8,6))
plt.boxplot([setosa ,versicolor,virginica] ,labels=["setosa" ,"versicolor","virginica"])
plt.title("Boxplot des longueurs des pétales par espèce")
plt.ylabel("petal length")
plt.xlabel("espace")
plt.show()
```

# Partie E — Visualisations Avancées

### E22. Subplots comparatifs
Deux scatter plots côte à côte : Setosa vs Versicolor.


```python
# Votre code ici
setosa_x = X_clean_id[Y == "Iris-setosa" , 2]
setosa_y = X_clean_id[Y == "Iris-setosa" , 3]
versicolor_x = X_clean_id[Y == "Iris-versicolor" , 2]
versicolor_y = X_clean_id[Y == "Iris-versicolor" , 3]
fig , (ax1 , ax2) = plt.subplots(1,2,figsize=(12,5))
ax1.scatter(setosa_x , setosa_y , color = "red" , label="Setosa")
ax1.set_title("setosa")
ax1.set_xlabel("petal length")
ax1.set_ylabel("petal width")
ax1.legend()

ax2.scatter(versicolor_x , versicolor_y , color = "red" , label="versicolor")
ax2.set_title("versicolor")
ax2.set_xlabel("petal length")
ax2.set_ylabel("petal width")
ax2.legend()
plt.show()
```
